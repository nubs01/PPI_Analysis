```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import easygui as eg
from ppi_analysis import *
```

```python
data_files = eg.fileopenbox(multiple=True,msg='Select all PPI data files')
```

```python
data = pd.DataFrame()
Nnan = 0
for file in data_files:
    tmp = import_ppi_data(file)
    Nnan = Nnan + data.isnull().any(axis=1).sum()
    tmp = tmp.dropna()
    data = data.append(tmp)
data.reset_index(inplace=True,drop=True)
```

```python
data['Gender'],data['Genotype'] = zip(*data['Group'].map(parse_group))
data = parse_session_and_trial(data)
```

```python
# grab rPPI data
ppiData = data.query('Session_Type=="rPPi"')
```

```python
ppiData["% PPI"] = np.nan
ppiData['% PPI'] = ppiData.apply(lambda x: calcPPI(x,ppiData),axis=1)
```
